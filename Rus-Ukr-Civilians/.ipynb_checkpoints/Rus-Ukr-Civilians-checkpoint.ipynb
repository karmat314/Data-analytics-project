```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load data
df = pd.read_csv('ukraine-civilian.csv')

# Clean Fatalities: remove '+' and convert to int
df['Fatalities'] = df['Fatalities'].astype(str).str.replace('+', '', regex=False)
df['Fatalities'] = pd.to_numeric(df['Fatalities'], errors='coerce').fillna(0).astype(int)

# Split Time Period into Start_Date and End_Date
df[['Start_Date', 'End_Date']] = df['Time Period'].str.split('–', expand=True)

# Strip whitespace
df['Start_Date'] = df['Start_Date'].str.strip()
df['End_Date'] = df['End_Date'].str.strip()

# Convert to datetime (keep pandas datetime)
df['Start_Date'] = pd.to_datetime(df['Start_Date'], dayfirst=True, errors='coerce')
df['End_Date'] = pd.to_datetime(df['End_Date'], dayfirst=True, errors='coerce')

# Remove time part by converting to date only
df['Start_Date'] = df['Start_Date'].dt.date
df['End_Date'] = df['End_Date'].dt.date

# Calculate Duration Days (add 1 to count both start and end days)
df['Duration_Days'] = (pd.to_datetime(df['End_Date']) - pd.to_datetime(df['Start_Date'])).dt.days + 1

# Normalize Fatalities and Duration_Days using MinMaxScaler
scaler = MinMaxScaler()
df[['Fatalities_Norm', 'Duration_Norm']] = scaler.fit_transform(df[['Fatalities', 'Duration_Days']])

# Display the cleaned DataFrame head
print(df[['Area', 'Fatalities', 'Start_Date', 'End_Date', 'Duration_Days', 'Fatalities_Norm', 'Duration_Norm']].head())

# Visualization 1: Fatalities by Area (bar plot)
plt.figure(figsize=(12,6))
plt.bar(df['Area'], df['Fatalities'], color='crimson')
plt.xticks(rotation=90)
plt.title('Fatalities by Area')
plt.xlabel('Area')
plt.ylabel('Fatalities')
plt.tight_layout()
plt.show()

# Visualization 2: Normalized Fatalities and Duration (scatter plot)
plt.figure(figsize=(10,6))
plt.scatter(df['Duration_Norm'], df['Fatalities_Norm'], color='navy')
plt.title('Normalized Fatalities vs. Duration Days')
plt.xlabel('Normalized Duration Days')
plt.ylabel('Normalized Fatalities')
plt.grid(True)
plt.show()
```
